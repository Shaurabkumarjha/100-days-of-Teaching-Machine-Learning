```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import scipy.stats as stats
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score

from sklearn.preprocessing import PowerTransformer
```

```python
df = pd.read_csv('concrete_data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
x= df.drop(columns=['Strength'])
y=df.iloc[:,:-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
# apply regression without any transformation
lr = LinearRegression()
lr.fit(x_train,y_train)

y_pre = lr.predict(x_test)
```

```python
r2_score(y_test,y_pre)
```

```python
# cross checking
lr = LinearRegression()
np.mean(cross_val_score(lr,x,y,scoring='r2'))
```

```python
# Ploting the distplot without any transfromation

for col in x_train.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(x_train[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(x_train[col],dist='norm',plot=plt)
    plt.show()
```

```python
# Applying Box-Cox Transform
pt=PowerTransformer(method='box-cox')
x_train_transformed=pt.fit_transform(x_train+0.000001)
x_test_transformed = pt.transform(x_test+0.000001)

pd.DataFrame({'cols':x_train.columns,'box_cox_lambdas':pt.lambdas_})
```

```python
x_train_transformed
```

```python
# apply regression without any transformation
lr = LinearRegression()
lr.fit(x_train_transformed,y_train)

y_pre1 = lr.predict(x_test_transformed)
r2_score(y_test,y_pre1)
```

```python
# cross checking
pt = PowerTransformer(method='box-cox')
x_trans = pt.fit_transform(x+0.000001)

lr = LinearRegression()
np.mean(cross_val_score(lr,x_trans,y,scoring='r2'))
```

```python
# Convert x_train_transformed to a DataFrame, using the same column names as x_train
x_train_transformed_df = pd.DataFrame(x_train_transformed, columns=x_train.columns)

for col in x_train.columns:
    plt.figure(figsize=(14, 4))
    
    # Plot original data distribution
    plt.subplot(121)
    sns.histplot(x_train[col], kde=True, stat="density")
    plt.title(f'Distribution of {col} in x_train')

    # Plot transformed data distribution
    plt.subplot(122)
    sns.histplot(x_train_transformed_df[col], kde=True, stat="density")
    plt.title(f'Distribution of {col} in x_train_transformed')

    plt.show()
```

```python
# Apply yeo Johnson transformer
pt1 = PowerTransformer()

x_train_tns = pt1.fit_transform(x_train)
x_test_tns = pt1.transform(x_test)

lr=LinearRegression()
lr.fit(x_train_tns,y_train)

y_pre = lr.predict(x_test_tns)

print(r2_score(y_test,y_pre))

pd.DataFrame({'cols':x_train.columns,'Yeo_Johnson_lambda':pt1.lambdas_})
```
