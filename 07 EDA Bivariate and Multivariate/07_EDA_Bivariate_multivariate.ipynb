```python
import pandas as pd
import seaborn as sns
```

```python
# this function is used for loading by default data of seaborn
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tit = pd.read_csv('train.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

# 1. Scatterplot (Numerical - Numerical)

```python
# Bivariate
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
# multivariate
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='smoker',style='sex',size='size')
```

# 2. Bar Plot (Numerical - categorical)

```python
tit.head()
```

```python
# Bivariate
sns.barplot(x='Pclass',y='Age',data=tit)
```

```python
# Multivariate
sns.barplot(x='Pclass',y='Age',data=tit,hue='Sex')
```

# 3. Box Plot (Numerical - Categorical)

```python
# Bivariate
sns.boxplot(x='Pclass',y='Age',data=tit)
```

```python
sns.boxplot(x='Pclass',y='Age',data=tit,hue='Sex')
```

# 4. Distplot (Numerical - Categoriacla)

```python
sns.distplot(tit[tit['Survived']==0]['Age'])
sns.distplot(tit[tit['Survived']==1]['Age'])
```

```python
sns.distplot(tit[tit['Survived']==0]['Age'],hist=False)
sns.distplot(tit[tit['Survived']==1]['Age'],hist=False)
```

# 5. HeatMap (Categorical - Categorical)

```python
tit.head(3)
```

```python
sns.heatmap(pd.crosstab(tit['Pclass'],tit['Survived']))
```

```python
(tit.groupby('Embarked')['Survived'].mean() * 100)
```

# 6. Cluster Map(Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(tit['Pclass'],tit['Survived']))
```
